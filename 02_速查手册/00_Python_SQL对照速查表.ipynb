## 🟢 第一部分：Python 内置基础

```python
# 1. 列表操作 (List)
sales = [100, 200, 150]

# 增 (Append)
sales.append(300)
print(f"追加后: {sales}")

# 删 (Pop)
last_one = sales.pop()
print(f"弹出了: {last_one}, 现在的列表: {sales}")

# 查 (Len & Sum)
print(f"个数: {len(sales)}, 总和: {sum(sales)}")
```

```python
# 2. 循环神器 (Zip)
days = ['Mon', 'Tue', 'Wed']
amounts = [10, 20, 30]

print("--- 开始 Zip 循环 ---")
for d, a in zip(days, amounts):
    print(f"{d} 卖了 {a} 元")
```

## 🔵 第二部分：Pandas 核心 (SQL 对照)

```python
import pandas as pd

# 创建一个 Mock 数据
df = pd.DataFrame({
    'city': ['Beijing', 'Shanghai', 'Beijing', 'Shenzhen', 'Shanghai'],
    'sales': [100, 200, 150, 300, 250],
    'date': ['Mon', 'Mon', 'Tue', 'Tue', 'Wed']
})

print("原始表格:")
df  # 在 Notebook 里直接写变量名就能展示
```

### 1. 筛选 (WHERE)
SQL: `SELECT * FROM table WHERE sales > 150`

```python
df[df['sales'] > 150]
```

### 2. 聚合 (GROUP BY)
SQL: `SELECT city, SUM(sales) FROM table GROUP BY city`

```python
df.groupby('city')['sales'].sum()
```
