## Drill 2: 脏字符大扫除 (Dirty String Scrubber) 🧹
**痛点**: `str.replace` 怎么用？正则怎么写？怎么一次换多个？
**口诀**: `Series.str.replace(r'正则', '', regex=True)`
**任务**: 清洗以下金额列，转成 float。

```python
# 练习 2.1: 去除 '$' 和 ','
df_money1 = pd.DataFrame({'Price': ['$1,000', '$2,500', '$500']})
# 你的代码:
df_money1['Price'] = pd.to_numeric(df_money1['Price'].str.replace(r'[$,]', '',regex=True))
df_money1
# 练习 2.2: 去除 'RM' (马来西亚币) 和 空格
df_money2 = pd.DataFrame({'Cost': ['RM 100', 'RM 200 ', 'RM 50']})
# 你的代码:
df_money2['Cost'] = pd.to_numeric(df_money2['Cost'].str.replace(r'[a-zA-Z ]','',regex=True))
df_money2
# 练习 2.3: 终极混合 (包含 'USD', '$', ',', 'approx')
df_money3 = pd.DataFrame({'Value': ['USD 1,000', '$500 approx', '200']})
# 你的代码 (提示: r'[USD$,a-z ]')
df_money3['Value'] = pd.to_numeric(df_money3['Value'].str.replace(r'[USD$,a-z ]','',regex=True))
df_money3

# 检查
print(df_money1)
print(df_money2)
print(df_money3)
```

## Drill 4: 逻辑筛选 (Logic Gate) 🚪
**痛点**: `&` 左右没加括号；变量名里有非法符号。
**任务**: 筛选出 "2024年" 且 "销量 > 100" 的行。

```python
df_sales = pd.DataFrame({
    'Date': pd.to_datetime(['2024-01-01', '2023-12-31', '2024-05-01']),
    'Sales': [150, 200, 50]
})

# 你的代码 (记得用括号包起来!)
result = (df_sales[
    (df_sales['Date'].dt.year == 2024) & (pd.to_numeric(df_sales['Sales']) > 100)
])

print(result)
```

## Drill 5: 综合大考 (The Final Boss) 👹

既然您问了，我就不客气了。
这是 **Pandas 清洗地狱难度**。这份数据包含了现实中所有的恶梦：
1.  **重复值** (`drop_duplicates`)
2.  **列不齐** (`rename`, `str.strip`, `str.lower`)
3.  **缺失值** (`dropna` vs `fillna`)
4.  **拆分列** (e.g., "Beijing, China" -> City, Country)
5.  **类型转换** ($88k -> 88000)

您的目标：把这堆垃圾变成金子。

```python
import numpy as np
import pandas as pd

# 0. 造一个超级脏的 DataFrame
chaos_data = {
    'User ID ': ['U001', 'U002', 'U001', 'U003', 'U004', 'U005 '], # 有ID重复，有后缀空格
    ' Gender': ['M', 'F', 'M', 'Female', 'Male', np.nan],       # 有前缀空格，性别不统一
    'Location': ['Beijing, CN', 'Shanghai, CN', 'Beijing, CN', 'Tokyo, JP', 'New York, US', 'London'], # 有没逗号的
    'Salary': ['$80k', '$120k', '$80k', '¥5.0M', 'UNK', '100000'], # 货币单位乱飞
    'Age': [25, 30, 25, 120, np.nan, 28]                         # 有异常值(120岁)，有缺失
}
df_chaos = pd.DataFrame(chaos_data)
print("---- 处理前的数据 ----")
print(df_chaos)

# --- 你的任务 ---

# 1. 去重 (Remove Duplicates based on User ID)
df_chaos_cleaned = df_chaos.drop_duplicates(subset='User ID ')
# 2. 清洗列名 (Remove whitespace, convert to lowercase, replace space with _)
df_chaos_cleaned.columns = df_chaos_cleaned.columns.str.strip().str.lower().str.replace(r' ','',regex=True)
# 3. 处理 Gender (把 'Female' -> 'F', 'Male' -> 'M', 缺失值填 'U')
df_chaos_cleaned['gender'] = df_chaos_cleaned['gender'].str.replace('Female','F').str.replace('Male','M').fillna('U')
# 4. 拆分 Location (Split into 'City' and 'Country', 如果没有逗号 Country 填 'Unknown')
df_chaos_cleaned[['city','country']]=df_chaos_cleaned['location'].str.split(',',expand=True)
# 5. 清洗 Salary (把 '$', 'k' 去掉，乘1000；把 'M' 去掉乘 1000000；非数字行删掉或变为 NaN)
def clean_amount(x):
    if isinstance(x,str):
        x = x.replace('$','').replace('¥','').replace('UNK','NaN')
        if 'k' in x:
            x = float(x.replace('k',''))*1000
        elif 'M' in x:
            x = float(x.replace('M',''))*1000000
        else:
            x = float(x)
    return x

df_chaos_cleaned['salary'] = pd.to_numeric(df_chaos_cleaned['salary'].apply(clean_amount))
# 6. 处理 Age (把 > 100 的当做 NaN，然后用平均年龄填充所有 NaN)
def clean_age(x):
    if isinstance(x,int):
        if x > 100:
            x = pd.nan 
        else:
            x = int(x)
    return x 

df_chaos_cleaned['age'] = pd.to_numeric(df_chaos_cleaned['age'].apply(clean_age))
mean_age = np.mean(df_chaos_cleaned['age'])
df_chaos_cleaned['age'] = df_chaos_cleaned['age'].fillna(mean_age)
print(df_chaos_cleaned)
```

## Drill 6: 拒绝 For 循环 (Vectorization Gym) 🚀

这里有 3 个最常见的 "Apply 陷阱"，您的任务是**不要用 apply**，全部用向量化操作搞定。

**工具箱提示**:
1.  `np.where(条件, 是, 否)`: 替代 `if...else`

3.  `.dt.year`: 替代 `datetime.year`
