## 1. 核心概念：向量化 (Vectorization) ⚡️

### ❌ Bad Practice (Loop)
想象你在 Excel 里，如果要把 A 列和 B 列相加，你会一行一行自己加吗？
Python 的 `for` 循环就是“一行一行自己加”，巨慢无比。

````python
import numpy as np
import time

# 创建 100万行数据
size = 1000000
a_list = list(range(size))
b_list = list(range(size))

# 🐢 方法 1: For Loop (极慢)
start = time.time()
result_list = []
for i in range(size):
    result_list.append(a_list[i] + b_list[i])
end = time.time()
print(f"For Loop 耗时: {end - start:.5f} 秒")
````

### ✅ Good Practice (NumPy Vectorization)
**Vectorization (向量化)** 就是 Excel 的“整列相加”。CPU 会并行处理这 100 万个数字。

````python
# 🐇 方法 2: NumPy (极快)
a_arr = np.array(a_list)
b_arr = np.array(b_list)

start = time.time()
result_arr = a_arr + b_arr  # 就像 Excel: Column A + Column B
end = time.time()
print(f"NumPy 耗时: {end - start:.5f} 秒")
````

## 2. 核心概念：广播 (Broadcasting) 📡

这是 NumPy 最难理解也最骚的操作。
**场景**: 你有一个 `(3, 3)` 的矩阵（比如 3 个学生，每人 3 门课的成绩）。
现在想给**每门课**都加分（数学+10, 英语+5, 体育+0）。

### 规则:
如果两个数组形状不同，NumPy 会自动把“小数组”**复制扩充**成“大数组”，然后对位计算。

````python
# 3个学生，3门课成绩
scores = np.array([
    [80, 90, 70],  # Student 1
    [60, 70, 80],  # Student 2
    [90, 95, 85]   # Student 3
])

# 加分政策: [Math+10, Eng+5, Gym+0]
bonus = np.array([10, 5, 0])

# ⚡️ Broadcasting 发生的地方
final_scores = scores + bonus
print("原始成绩:\n", scores)
print("加分后:\n", final_scores)

# 思考: 为什么 (3,3) 可以和 (3,) 相加？
# 答案: NumPy 自动把 bonus 复制了 3 行，变成了 (3,3) 再相加。
````

## 3. 统计学实战: A/B Test 的基石 📊

> **🤔 疑问**: "很多函数 Pandas 里也有，为什么还要学 NumPy 的版本？"
>
> **答案**:
> 1.  **Pandas 是壳，NumPy 是芯**: `df.mean()` 底下调用的其实就是 `np.mean()`。
> 2.  **机器学习只认 NumPy**: 稍后我们学的 Scikit-Learn (预测模型) **不认识 Pandas 的列名**，只认识纯数字的 NumPy 矩阵。
> 3.  **速度**: 对于超大规模运算，NumPy 直接操作内存，比 Pandas 的封装少了一层，速度更快。

面试里常问: **如何计算方差 (Variance) 和 标准差 (Standard Deviation)?**
不要背公式 $\frac{\sum (x - \mu)^2}{N}$。

用 NumPy，一行代码解决。

*   **Mean (均值)**: `np.mean()` —— 这一组数据的平均水平。
*   **Std (标准差)**: `np.std()` —— 这一组数据由于“运气”导致的波动有多大？
    *   **Std 大**: 说明数据很不稳定（比如币圈）。
    *   **Std 小**: 说明数据很稳定（比如国债）。

````python
# 模拟两组 A/B Test 数据 (转化率)
# Group A: 比较稳定
group_a = np.array([0.10, 0.11, 0.10, 0.09, 0.10])

# Group B: 虽然均值一样，但波动极大
group_b = np.array([0.01, 0.20, 0.01, 0.19, 0.09])

print(f"Group A Mean: {np.mean(group_a):.2f}, Std: {np.std(group_a):.2f}")
print(f"Group B Mean: {np.mean(group_b):.2f}, Std: {np.std(group_b):.2f}")

# 💡 面试题: 如果两组均值一样，但 B 组的标准差很大，你会选哪个？
# 答案: 选 A。B 组太不稳定，风险高，可能是辛普森悖论或者异常值导致的虚高。
````

### 练习 2: 谁是离群点？ (Filter & Stats)
在一组成绩中，找出所有 "偏离平均值超过 2 倍标准差" 的异常分数。

```python
scores = np.array([80, 85, 90, 88, 82, 10, 85, 95, 120, 84])
```

**需求**:
1. 计算 `mean` 和 `std`。
2. 找出所有 `x > mean + 2*std` 或 `x < mean - 2*std` 的数字。

````python
# ✍️ 在这里写代码
scores = np.array([80, 85, 90, 88, 82, 10, 85, 95, 120, 84])

stats_mean = np.mean(scores)
stats_std = np.std(scores)

# Upper Bound & Lower Bound
filter_1 = stats_mean + 2*stats_std
filter_2 = stats_mean - 2*stats_std

# Boolean Indexing
scores_filter = scores[(scores > filter_1)|((scores < filter_2))]
scores_filter
````

### 💡 Pro Tip: 高手怎么写？(Absolute Value)

你的写法非常直观，**逻辑完全正确**！💯
但在 NumPy 工业界，我们通常喜欢用 **`np.abs` (绝对值)** 来偷懒。

因为 "离群" 意味着 "距离平均值太远"，不管是正向远还是负向远，**距离 (绝对值)** 是一样的。

公式: $|x - \mu| > 2\sigma$

````python
# 🚀 一行代码流
condition = np.abs(scores - np.mean(scores)) > 2 * np.std(scores)
pro_outliers = scores[condition]
pro_outliers
````
