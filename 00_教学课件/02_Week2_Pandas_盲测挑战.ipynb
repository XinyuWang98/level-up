```python
import pandas as pd
import numpy as np

# 🛒 数据加载 (这次我们模拟一个复杂一点的员工绩效表)
data = {
    'emp_id': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
    'dept': ['Sales', 'Tech', 'Tech', 'Sales', 'HR', 'Tech', 'Sales', 'HR'],
    'city': ['Beijing', 'Shanghai', 'Beijing', 'Shenzhen', 'Beijing', 'Shanghai', 'Shenzhen', 'Shanghai'],
    'salary': [12000, 25000, 18000, 11000, 9000, 26000, 13000, 9500],
    'performance': [1.2, 0.9, 1.1, 1.5, 1.0, 0.8, 1.3, 1.0],  # 绩效系数
    'age': [25, 30, 28, 24, 26, 32, 25, 27]
}

df = pd.DataFrame(data)
print("盲测数据准备就绪！")
df
print("===info===")
print(df.info())
print("===describe===")
print(df.describe())
```

---

### ⚔️ 挑战 1: 筛选 (Filter)
找出所有 **“Tech”部门** 且 **工资大于 20000** 的员工。

```python
# [你的代码]
emp_dept_salary = df.groupby(['dept','emp_id'])['salary'].sum().reset_index()
fliter_tech_over_2w = emp_dept_salary[(emp_dept_salary['dept']=='Tech')&(emp_dept_salary['salary']>=20000)]

fliter_tech_over_2w
```

### ⚔️ 挑战 2: 新增列 (Vectorization)
计算每个人的 **年终奖 (bonus)**。
*   公式：`bonus = salary * performance`
请直接把这个新列加到 `df` 里。

```python
# [你的代码]
df['bonus']=df['salary']*df['performance']
df
```

### ⚔️ 挑战 3: 分组聚合 (GroupBy)
在这个公司里，**哪个部门最烧钱？**
请按部门 (`dept`) 分组，计算每个部门的：
1.  总工资 (`sum`)
2.  平均绩效系数 (`mean`)

要求：结果必须是一个平铺的表格 (提示：重置索引)。

```python
# [你的代码]
# ddept_salary=df.groupby('dept')['salary'].sum().reset_index()
# dept_mean_performance = df.groupby('dept')['performance'].mean().reset_index()
dept_stats = df.groupby('dept').agg({
    'salary':'sum',
    'performance':'mean'
}).reset_index()
# dept_stats
df_merge = pd.merge(df,dept_stats,on='dept',suffixes=('','_dept_avg'))
df_merge
```

### ⚔️ 挑战 4: 排序 (Sorting)
谁是**“年终奖之王”**？
请按你刚才算出来的 `bonus` 对全表进行 **降序** 排序，并只显示前 3 名。

```python
# [你的代码]
# 找出 bonus 最大的前 3 名
df.sort_values(by='bonus', ascending=False).head(3)
```

### ⚔️ 挑战 5: 窗口函数 (Window Logic)
**（附加题）**
每个人都想知道：**“我的工资在咱们部门里排第几？”**

请新增一列 `dept_rank`，计算每个人在**自己部门内**的**工资排名** (从高到低)。
*   提示：`groupby` 之后不仅可以接 `agg` / `transform`，还可以接 `rank` 哦！尝试推理一下写法。

```python
# [你的代码]
df['dept_salary_rank']=df.groupby('dept')['salary'].rank(method='dense',ascending=False)
df
```
