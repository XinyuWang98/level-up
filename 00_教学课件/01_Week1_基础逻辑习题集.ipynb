### 🟡 Level 2：计算全站客单价 (AOV)
*   **业务问题**：过去5天的**平均客单价**是多少？
*   **公式**：总金额 (Total GMV) / 总单数 (Total Orders)
*   **易错点**：千万别用 `sum(daily_revenue) / len(orders_log)`，那是日均营收！

```python
# ✏️ 在这里写你的代码

# 1. 算总金额

# 2. 算总单数 (难点!)

# 3. 除一下
```

```python
# 🔑 参考答案

total_gmv = sum([sum(day) for day in orders_log])  # 列表推导式写法，相当于把 Level 1 压缩成一行

total_orders = 0
for day_orders in orders_log:
    total_orders += len(day_orders)

if total_orders > 0:
    aov = total_gmv / total_orders
    print(f'总金额: {total_gmv}, 总单数: {total_orders}')
    print(f'全站客单价: {aov:.2f}')
else:
    print('暂无订单')
```
