## 🛠️ 准备活动：生成脏数据
不做任何修改，直接运行生成一份“糟糕”的数据。

```python
import pandas as pd
import numpy as np

data = {
    'date': ['2023-01-01', '01/02/2023', '2023.01.03', 'invalid_date', '2023-01-05'],
    'category': ['A', 'A', 'B', 'B', 'C'],
    'amount': ['$100', '200', '300.5', '400', '500'],
    'user': ['u1', 'u1', 'u2', 'u2', 'u3']
}

df = pd.DataFrame(data)
df
```

## ⚔️ 第一组训练：日期解析 (Date Parsing)
**痛点回忆**：默认解析会报错，或者解析不全。
**目标**：把 `date` 列转为标准日期格式 `YYYY-MM-DD`，处理不了的变成 `NaT`。

```python
# [👉 请在这个单元格写代码]
# 提示：pd.to_datetime(..., format='mixed', errors='coerce')
df['date'] = pd.to_datetime(df['date'], format='mixed', errors='coerce')
df
```

## ⚔️ 第二组训练：数值回魂 (Numeric Conversion)
**痛点回忆**：`$100` 是字符串，没法求和。
**目标**：
1. 去掉 `amount` 列里的 `$` 符号。
2. 把它转为 `float` 类型。

```python
# [👉 请在这个单元格写代码]
df['amount'] = df['amount'].str.replace('$','').astype(float)
df
```

## ⚔️ 第三组训练：分组聚合 (Group By)
**痛点回忆**：**List/Set 不是 DataFrame**。要用 `groupby` 这种“魔法”来变出新表。
**目标**：
按 `user` 分组，算出：
1. `sum`: 总金额
2. `count`: 消费次数

最后生成一个新的 `user_stats` DataFrame。

```python
# [👉 请在这个单元格写代码]
# 核心口诀：df.groupby('谁')['算谁'].agg(['怎么算'])
user_stats = df.groupby('user')['amount'].agg(['sum', 'count']).reset_index()

user_stats
```
