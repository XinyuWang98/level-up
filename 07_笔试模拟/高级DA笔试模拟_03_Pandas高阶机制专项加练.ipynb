### 1.1 找出活跃度达标的“全团成员” (groupby + filter)
**业务场景**：一个游戏公会（Guild）里有多个玩家。如果某个公会的**平均活跃天数 >= 10天**，则该公会达标，我们需要提取出该达标公会里的**所有玩家（哪怕有人的活跃度只有 1 天）**。
*预期结果：只保留 Guild 为 A 的全部 3 名玩家，剔除 Guild B 所有人。*

```
import pandas as pd
import numpy as np
df_guild = pd.DataFrame({
    'guild': ['A', 'A', 'A', 'B', 'B'],
    'player': ['P1', 'P2', 'P3', 'P4', 'P5'],
    'active_days': [12, 15, 5, 2, 8]
})
# 你的代码
```

```
# ✅ 参考答案
res = df_guild.groupby('guild').filter(lambda x: x['active_days'].mean() >= 10)
display(res)
```

### 1.2 找出包含“作弊玩家”的整个对局 (groupby + filter) 
**业务场景**：一旦某个对局（Match_ID）中出现任何一个标记为作弊（is_cheat=1）的玩家，我们就要把这个对局里的**所有 10 个人全部拉出来排查**。
*预期结果：Match_ID 1001 中 P2 作弊，所以 P1和P2 都要被拉出来；1002 没人作弊，全员安全不显示。*

```
import pandas as pd
import numpy as np
df_match = pd.DataFrame({
    'match_id': [1001, 1001, 1002, 1002],
    'player': ['P1', 'P2', 'P3', 'P4'],
    'is_cheat': [0, 1, 0, 0]
})
# 你的代码
```

```
# ✅ 参考答案
# 💡 逻辑：只要组里 is_cheat 的最大值 == 1 (或者 sum > 0)，就说明有人作弊
res = df_match.groupby('match_id').filter(lambda x: x['is_cheat'].max() == 1)
display(res)
```

### 2.1 财务账单分箱的边界陷阱 (pd.cut include_lowest)
**业务场景**：按照订单金额分类：
- 0 - 50 元：Low
- 50 - 100 元：High
有一个订单恰好是 **0 元** (比如用了白嫖券)，如果不加特定参数，它会变成 NaN。
请用 `pd.cut` 并确保 0 元订单被正确分入 'Low' 组。

```
import pandas as pd
import numpy as np
df_order = pd.DataFrame({'amt': [0, 25, 50, 80]})
# bins 必须是 [0, 50, 100]
# 你的代码
```

```
# ✅ 参考答案
df_order['level'] = pd.cut(df_order['amt'], bins=[0, 50, 100], labels=['Low', 'High'], include_lowest=True)
display(df_order)
```

### 3.1 跨月订阅到期计算 (DateOffset)
**业务场景**：用户在 2024年1月31日 购买了 1 个月的会员。
请算出到期日。注意：如果直接加 30 天，闰年和平年的2月会算错；必须让他智能跳到 2月29日（2024是闰年）。

```
import pandas as pd
import numpy as np
df_sub = pd.DataFrame({'buy_date': pd.to_datetime(['2024-01-31'])})
# 你的代码 (使用 pd.DateOffset)
```

```
# ✅ 参考答案
df_sub['expire_date'] = df_sub['buy_date'] + pd.DateOffset(months=1)
display(df_sub)
```

### 4.1 动态名单过滤 (Query + @)
**业务场景**：找出部门在 `target_departments` 列表中，**并且** 评级等于 `target_level` 的员工。
请**只用一行 `.query()`** 来完成，不要用传统的 `df[(df...in...) & (df...==...)]`

```
import pandas as pd
import numpy as np
df_emp = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'dept': ['Sales', 'Tech', 'HR', 'Tech'],
    'level': ['P5', 'P6', 'P5', 'P5']
})
target_departments = ['Tech', 'Sales']
target_level = 'P5'

# 你的代码 (只写一行 query)
```

```
# ✅ 参考答案
res = df_emp.query("dept in @target_departments and level == @target_level")
display(res)
```
